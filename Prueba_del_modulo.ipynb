```python
# cargar los datos de Seaborn
iris = sns.load_dataset("iris")
iris.head()
```

```python
# extraer matriz de caracter√≠sticas y vector de etiquetas
X = iris.values[:100,2:4]
y = iris.values[:,-1]
iris.species.replace(['setosa','versicolor','virginica'],[0,1,2], inplace=True) #cambiar a etiquetas numericas
y=iris.values[:100,-1].astype('int32')
X.shape, y.shape
```

```python
#normalizar con sklearn
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_norm = scaler.fit_transform(X)
```

```python
#separar conjunto de datos
X_train, X_test, y_train, y_test = ml.SplitData(X_norm, y, 0.3)
```
